```python
%%time
import keras 
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten
from keras.layers import Conv2D,MaxPooling2D
from keras.optimizers import RMSprop
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np


def def_model(in_shape, nb_classes):
    model = Sequential()
    model.add(Conv2D(32,
              kernel_size = (3, 3),
              activation = 'relu',
              input_shape = in_shape))
    model.add(Conv2D(32, (3, 3), activation = 'relu'))
    model.add(MaxPooling2D(pool_size = (2, 2)))
    model.add(Dropout(0.25))

    model.add(Conv2D(64, (3, 3), activation = 'relu'))
    model.add(Conv2D(64, (3, 3), activation = 'relu'))
    model.add(MaxPooling2D(pool_size = (2, 2)))
    model.add(Dropout(0.25))

    model.add(Flatten())
    model.add(Dense(512, activation = 'relu'))
    model.add(Dropout(0.5))
    model.add(Dense(nb_classes, activation = 'softmax'))
    model.summary()
    return model

def get_model(in_shape, nb_classes):
    model = def_model(in_shape, nb_classes)
    model.compile(
        loss = 'categorical_crossentropy',
        optimizer = RMSprop(),
        metrics = ['accuracy'])
    return model

# 変更必要
im_size = 64
im_color = 1
in_shape = (im_size, im_size, im_color)
nb_classes = 20 

#変更必要
photos = np.load('D:/過去の研究データ/2023npz/2023UsingFrontAndBackHands/2023Num20EdgesKarnel5-64pix.npz')
x = photos['x']
y = photos['y']

x = x.reshape(-1, im_size, im_size, im_color)
x = x.astype('float32') / 255
y = keras.utils.to_categorical(y.astype('int32'), nb_classes)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, train_size = 0.9)

model = get_model(in_shape, nb_classes)

#変更必要
hist = model.fit(x_train, y_train,
          batch_size = 256, 
          epochs = 20,
          verbose = 1,
          validation_data = (x_test, y_test))

score = model.evaluate(x_test, y_test, verbose = 1)
print('正解率=', score[1], 'loss=', score[0])


plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('Accuracy')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()

plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('Loss')
plt.legend(['train', 'test'], loc = 'upper left')
plt.show()

#変更必要
model.save_weights('D:/過去の研究データ/2023hdf/2023UsingFrontAndBackHands/2023Num20EdgesKarnel5-64pixBatch256.hdf5') 
```

```python
#他人受入率評価用のプログラム 
import keras 
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten
from keras.layers import Conv2D,MaxPooling2D
from keras.optimizers import RMSprop
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image


def def_model(in_shape, nb_classes):
    model = Sequential()
    model.add(Conv2D(32,
              kernel_size = (3, 3),
              activation = 'relu',
              input_shape = in_shape))
    model.add(Conv2D(32, (3, 3), activation = 'relu'))
    model.add(MaxPooling2D(pool_size = (2, 2)))
    model.add(Dropout(0.25))

    model.add(Conv2D(64, (3, 3), activation = 'relu'))
    model.add(Conv2D(64, (3, 3), activation = 'relu'))
    model.add(MaxPooling2D(pool_size = (2, 2)))
    model.add(Dropout(0.25))

    model.add(Flatten())
    model.add(Dense(512, activation = 'relu'))
    model.add(Dropout(0.5))
    model.add(Dense(nb_classes, activation = 'softmax'))
    return model


def get_model(in_shape, nb_classes):
    model = def_model(in_shape, nb_classes)
    model.compile(
        loss = 'categorical_crossentropy',
        optimizer = RMSprop(),
        metrics = ['accuracy'])
    return model

#他人受入率用のテスト配列
omote = ["gotou/omote",
        "isii/omote",
        "mimura/omote",
        "nakamura/omote",
        "okutani/omote",
        "saitouhide/omote",
        "saitouryou/omote",
        "sakuma/omote",
        "tanaka/omote",
        "tochikubo/omote"]
ura = ["gotou/ura",
     "isii/ura",
     "mimura/ura",
     "nakamura/ura",
     "okutani/ura",
     "saitouhide/ura",
     "saitouryou/ura",
     "sakuma/ura",
     "tanaka/ura",
     "tochikubo/ura"]

laplace5 = [[-1, -3, -4, -3, -1],
           [-3, 0, 6, 0, -3],
           [-4, 6, 20, 6, -4],
           [-3, 0, 6, 0, -3],
           [-1, -3, -4, -3, -1]]
#表の評価画像と裏の評価画像のパスが格納
target_omote = []
target_ura = []
#誤って他人を受入れてしまった確率が格納
passpercent = []
#正しく他人を拒否した確率が格納
notpasspercent = []
#閾値未満ではじかれた確率が表と裏でそれぞれ格納
omoteNotPercent = []
uraNotPercent = []
#閾値の設定⇒設定必要
thresh = 99.99999

#target_omoteに表の画像を入れ、target_uraに裏の画像を入れる
#変更必要
for k in range(10):
    for i in range(10):
        target_omote.append("D:/過去の研究データ/graytest/" + str(omote[k]) + "/" + str(i) + ".jpg")
        target_ura.append("D:/過去の研究データ/graytest/" + str(ura[k]) + "/" + str(i) + ".jpg")
        
#変更必要
im_size = 64
im_color = 1
in_shape = (im_size, im_size, im_color)
nb_classes = 20
#配列の最大値
max_score = 0.0

LABELS = ["相澤さん", "新井さん", "江波戸", "長谷川さん",
          "樋口さん", "石井さん", "小島さん", "中村さん",
          "大野さん", "坂井さん", "関根さん", "島田さん",
          "清水さん", "高橋さん", "田中ひろきさん", "田中みゆさん",
          "田中たいきさん", "徳山さん", "渡邉さん",  "山口さん"]

model = get_model(in_shape, nb_classes)
#変更必要
model.load_weights('D:/過去の研究データ/2023hdf/2023UsingFrontAndBackHands/2023Num20EdgesKarnel5-64pixBatch256.hdf5')

#評価画像の確率を受け取り、max_scoreよりも大きい値をmax_scoreとして更新する
def findmaxscore(per):
    global max_score
    if per != 100.0:
        if max_score <= per:
            max_score = per 
            
def check_photo(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    karnel = np.array(laplace5, np.float64)
    img = cv2.filter2D(img, -1, karnel)
    #画像のリサイズ
    result_img = cv2.resize(img, (im_size, im_size)) 
    plt.imshow(result_img, cmap = 'gray')
    plt.show()

    x = np.asarray(result_img)
    x = x.reshape(-1, im_size, im_size, im_color)
    x = x / 255
    
    #画像の予測
    pre = model.predict([x])[0]
    idx = pre.argmax()
    per = (pre[idx] * 100)
    #閾値(この段階では認証はしないが閾値未満の画像を格納する)
    if per < thresh and ("omote" in path): 
        omoteNotPercent.append(per)
    elif per < thresh and ("ura" in path):
        uraNotPercent.append(per)
    else:
        pass
    return (idx, per)

def check_photo_str(path_omote, path_ura):
    
    idx_omote, per_omote = check_photo(path_omote)
    idx_ura, per_ura = check_photo(path_ura)
    
    #表の評価画像と裏の評価画像が閾値以上かつ認証結果の人が同じでないと認証されない
    #Ex,表裏どちらも閾値以上でも表の手がAさん、裏の手がBさんだと認証されない
    if per_omote >= thresh and per_ura >= thresh and LABELS[idx_omote] == LABELS[idx_ura]:
        print("この手は表:", LABELS[idx_omote], " 裏:", LABELS[idx_ura], "です")
        print("可能性は表の手:", per_omote, "% 裏の手:", per_ura, "%なので認証します")
        
        #他人受入率に値するので他人受入率の配列に格納する
        passpercent.append(per_omote)
        passpercent.append(per_ura)
        
        #max_scoreよりも大きい確率かどうかを判定する
        findmaxscore(per_omote)
        findmaxscore(per_ura)
        
    else:
        print('\033[31m' + "可能性が表の手:", per_omote, "% 裏の手:", per_ura, "%なので認証できません" + '\033[0m')
        print('\033[31m' + "認証の結果は表の手:", LABELS[idx_omote], " 裏の手:", LABELS[idx_ura], "でした" + '\033[0m')
        #正しく他人を拒否した確率を配列に格納する
        notpasspercent.append(per_omote)
        notpasspercent.append(per_ura)
    
if __name__ == '__main__':
    for i in range(100):
        print("\nここからは" + str(i) + "番目の写真です")
        #check_photo_strに接続
        #表と裏の評価画像を渡して認証を行う
        check_photo_str(target_omote[i], target_ura[i])
        if i % 10 == 9:
            print("================================================")
            print("テスト対象の人が変わります")
            print("================================================")

    #他人受入率の配列の中身を表示
    print(passpercent)
    #他人拒否率の配列の中身を表示
    print(notpasspercent)
    print("他人受入に用いる個数" + str(len(passpercent)))
    #全体の他人受入率算出用に変数FARに他人受入の個数を格納
    FAR = len(passpercent)
    print("閾値未満で識別したテスト数" + str(len(notpasspercent)))
    print("表の手だけで閾値未満で識別したテスト数" + str(len(omoteNotPercent)))
    print("裏の手だけで閾値未満で識別したテスト数" + str(len(uraNotPercent)))
    #表裏合計、それぞれの他人受入(拒否)率を算出する
    totalpass = len(passpercent) / 200
    totalnotpass = len(notpasspercent) / 200
    omotenotpass = len(omoteNotPercent) / 100
    omotepass = (100 - len(omoteNotPercent)) / 100
    uranotpass = len(uraNotPercent) / 100
    urapass = (100 - len(uraNotPercent)) / 100
    print("================================================")
    #未学習の手を閾値未満で識別した割合
    print("他人拒否率" + str(round(totalnotpass * 100, 4)) + "%")
    #未学習の手を閾値以上で識別した割合
    print("他人受入率" + str(round(totalpass * 100, 4)) + "%")
    print("表の手だけで未学習の手を閾値未満で識別した割合" + str(round(omotenotpass * 100, 4)) + "%")
    print("表の手だけで未学習の手を閾値以上で識別した割合" + str(round(omotepass * 100, 4)) + "%")
    print("裏の手だけで未学習の手を閾値未満で識別した割合" + str(round(uranotpass * 100, 4)) + "%")
    print("裏の手だけで未学習の手を閾値以上で識別した割合" + str(round(urapass * 100, 4)) + "%")
    print("閾値以上の最大値は" + str(max_score)) 
    try:
        print("閾値以上の最小値は" + str(min(passpercent))) 
    except:
        pass
```
